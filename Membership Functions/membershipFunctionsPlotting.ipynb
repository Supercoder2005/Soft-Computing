```python
import numpy as np
import matplotlib.pyplot as plt
```

# Triangular Membership Function

```python
def triangular_MF(x,a,b,c):
    return np.maximum(0,np.minimum((x-a)/(b-a),(c-x)/(c-b)))
x = np.linspace(0,10,1000)
y = triangular_MF(x,2,5,8)
plt.plot(x,y)
plt.xlabel('Universe of Discourse')
plt.ylabel('Membership Values')
plt.grid(True)
plt.show()
```

```python
def triangular_MF(x,a,b,c):
    mu = []
    for value in x:
        if value <= a:
            mu.append(0)
        elif a < value <= b:
            mu.append((value-a)/(b-a))
        elif b < value <= c:
            mu.append((c-value)/(c-b))
        else:
            mu.append(0)
    return np.array(mu)

x = np.linspace(0,10,1000)
y = triangular_MF(x,1,4,7)

plt.plot(x,y)
plt.title("Triangular Membership Function")
plt.xlabel("Universe of Discourse")
plt.ylabel("Membership Values")
plt.grid(True)
plt.show()
```

# Trapezoidal Membership Function

```python
def trapezoidal_mf(x,a,b,c,d):
    mu = []
    for value in x:
        if value <= a:
            mu.append(0)
        elif a < value <= b:
            mu.append((value-a)/(b-a))
        elif b < value <= c:
            mu.append(1)
        elif c < value <= d:
            mu.append((d-value)/(d-c))
        else:
            mu.append(0)
    return np.array(mu)

x = np.linspace(0,10,100)
y = trapezoidal_mf(x,2,4,6,8)
plt.plot(x,y)
plt.xlabel('Universe of Discourse')
plt.ylabel('Membership Values')
plt.title("Trapezoidal Membership Function")
plt.grid(True)
plt.show()
```

# Gaussian Membership Function

```python
def gaussian_MF(x,mean,sigma):
    return np.exp(-((x-mean)**2/2*sigma**2))
x = np.linspace(0,10,1000)
y = gaussian_MF(x,5,1)
plt.plot(x,y)
plt.title("Gaussian Membership Function")
plt.xlabel("Universe of Discourse")
plt.ylabel("Membership Value")
plt.grid(True)
plt.show()
```

# Generalized Bell Membership Function

```python
def generalizedBell_mf(x,c,a,b):
    return 1/(1+ np.abs((x-c)/a)**(2*b))
x = np.linspace(0,10,1000)
y = generalizedBell_mf(x,5,3,2)
plt.plot(x,y)
plt.title("Generalized Bell Membership Function")
plt.xlabel("Universe of Discourse")
plt.ylabel("Membership Value")
plt.grid(True)
plt.show()
```

# Sigmoidal Membership Function

```python
def sigmoidal_mf(x,a,c):
    return 1/(1+np.exp(-a*(x-c)))
x = np.linspace(0,10,1000)
y = sigmoidal_mf(x,2,5)
plt.plot(x,y)
plt.xlabel("Universe of Discourse")
plt.ylabel("Membership Value")
plt.title("Sigmoidal Membership Function")
plt.grid(True)
plt.show()
```
