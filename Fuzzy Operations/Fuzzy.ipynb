# Union of two Fuzzy sets

```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
def triangular_mf(x,a,b,c):
    mu = []
    for value in x:
        if value <= a:
            mu.append(0)
        elif a < value <= b:
            mu.append((value - a)/(b-a))
        elif b < value <= c:
            mu.append((c-value)/(c-b))
        else:
            mu.append(0)
    return np.array(mu)
```

```python
x = np.linspace(0,10,1000)
set1 = triangular_mf(x,1,3,5)
set2 = triangular_mf(x,2,4,6)
union = np.maximum(set1,set2)
plt.plot(x,set1)
plt.plot(x,set2)
plt.plot(x,union)
plt.xlabel("Universe of Discourse")
plt.ylabel("Membership Value")
plt.title("Union")
plt.fill_between(x,union,color="magenta",alpha=0.2)
plt.grid(True)
plt.show()
```

# Intersection of two Fuzzy sets

```python
intersection = np.minimum(set1,set2)
plt.plot(x,set1)
plt.plot(x,set2)
plt.plot(x,intersection)
plt.xlabel("Universe of Discourse")
plt.ylabel("Membership Value")
plt.title("Intersection")
plt.fill_between(x,intersection,color="magenta",alpha=0.2)
plt.grid(True)
plt.show()
```

# Complement of Fuzzy set

```python
complement = 1-set1 
plt.plot(x,set1)
plt.plot(x,complement)
plt.xlabel("Universe of Discourse")
plt.ylabel("Membership Value")
plt.title("Complement")
plt.fill_between(x,complement,color="magenta",alpha=0.2)
plt.grid(True)
plt.show()
```

# Algebric Sum

```python
algebric_sum = set1 + set2 - (set1*set2)
plt.plot(x,set1)
plt.plot(x,set2)
plt.plot(x,algebric_sum)
plt.xlabel("Universe of Discourse")
plt.ylabel("Membership Value")
plt.title("Algebraic Sum")
plt.fill_between(x,algebric_sum,color="magenta",alpha=0.2)
plt.grid(True)
plt.show()
```

# Algebraic Difference

```python
algebraic_diff = set1*(1-set2)
plt.plot(x,set1)
plt.plot(x,set2)
plt.plot(x,algebraic_diff)
plt.xlabel("Universe of Discourse")
plt.ylabel("Membership Value")
plt.title("Algebraic Diffrence")
plt.fill_between(x,algebraic_diff,color="magenta",alpha=0.2)
plt.grid(True)
plt.show()
```

# Cartesian Product

```python
R = np.array([[0.5,0.7,0.2],
              [0.8,0.6,0.9]])
S = np.array([[0.6,0.3],
              [0.9,0.4],
              [0.5,0.8]])
result = []
for i in range(len(R)):
    for j in range(len(R[0])):
        row = []

        for k in range(len(S)):
            for l in range(len(S[0])):

                value = min(R[i][j],S[k][l])
                row.append(value)
        result.append(row)
print(np.array(result))
```

# Max-Min Composition

```python
R = np.array([[0.5,0.7,0.2],
              [0.8,0.6,0.9]])
S = np.array([[0.6,0.3],
              [0.9,0.4],
              [0.5,0.8]])
final = []
for i in range(len(R)):
    result = []
    for j in range(len(S[0])):
        elementlist = []
        for k in range(len(S)):
            element = np.minimum(R[i][k],S[k][j])
            elementlist.append(element)
        result.append(max(elementlist))
    final.append(result)
print(np.array(final))
```

# Max-Product

```python
R = np.array([[0.5,0.7,0.2],
              [0.8,0.6,0.9]])
S = np.array([[0.6,0.3],
              [0.9,0.4],
              [0.5,0.8]])
final = []
for i in range(len(R)):
    result = []
    for j in range(len(S[0])):
        elementlist = []
        for k in range(len(S)):
            element = R[i][k]*S[k][j]
            elementlist.append(element)
        result.append(max(elementlist))
    final.append(result)
print(np.array(final))
```
